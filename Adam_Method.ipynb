# Nomor 11 (Adam Bashworth (metode Adam eksplisit) order 5 )


## Bagian A

```python
#defined variable
def adam_bashworth_ordinary_differential_equations_single_order_1(dt=0.02, pr=2):
    f=lambda t,y : 2-y
    f_t=lambda t :2+3*np.exp(-t)
    H = int((1-0)/dt)
    t=np.linspace(0, 1, H+1)
    y_t=np.zeros(len(t))
    y=np.zeros(len(t))
    e = np.zeros(len(t))
    y[0]=5
    y_t[0]=5
    e[0]=(np.abs(y[0]-y_t[0])/y_t[0])*100
    #looping
    for j in range(len(t)-1):
        # calculate y true
        y_t[j+1]=f_t(t[j+1])
        #calculate y_pred
        ## j+1=1
        if j+1==1:
          y[j+1]=y[j]+f(t[j], y[j])*dt

        ## j+1=2
        elif j+1==2:
          y[j+1]=y[j]+(3/2*f(t[j], y[j])-1/2*f(t[j-1], y[j-1]))*dt

        ## j+1=3
        elif j+1==3:
          y[j+1]=y[j]+(23/12*f(t[j], y[j])-16/12*f(t[j-1], y[j-1])\
                       +5/12*f(t[j-2], y[j-2]))*dt

        ## j+1=4
        elif j+1==4:
          y[j+1]=y[j]+(55/24*f(t[j], y[j])-59/24*f(t[j-1], y[j-1])\
                       +37/24*f(t[j-2], y[j-2])-9/24*f(t[j-3], y[j-3]))*dt

        ## j+1>= 5
        else:
          y[j+1]=y[j]+(191/72*f(t[j], y[j])-281/72*f(t[j-1], y[j-1])\
                       +267/72*f(t[j-2], y[j-2])-131/72*f(t[j-3], y[j-3])+\
                       26/72*f(t[j-4], y[j-4]))*dt



        #calculate error
        e[j+1]=(np.abs(y[j+1]-y_t[j+1])/y_t[j+1])*100
    df = pd.DataFrame(list(zip(t, y, y_t, e)),
                  columns =['t','y pred', 'y true', 'error(%)'])
    if pr == 1:
      return y_t, y, t
    else:
      display(df)
      plt.plot(t, y, '0.8', label='Euler Method')
      plt.plot(t, y_t, 'k', label='Analytical')
      plt.title('Comparison For Euler Method and Analytical')
      plt.xlabel('t')
      plt.ylabel('f(t)')
      plt.grid()
      plt.legend(loc='lower right')
      plt.show()
```

# Nomor 12 (Adam Bashworth Moulton)

## Bagian A

```python
#defined variable
def adam_bashworth_moulton(dt=0.1, pr=1, er=0.01):
    f=lambda t,y : 2-y
    f_t=lambda t :2+3*np.exp(-t)
    H = int((1-0)/dt)
    t=np.linspace(0, 1, H+1)
    y_t=np.zeros(len(t))
    y=np.zeros(len(t))
    e = np.zeros(len(t))
    k = np.zeros(len(t))
    y[0]=5
    y_t[0]=5
    e[0]=(np.abs(y[0]-y_t[0])/y_t[0])*100
    #looping
    for j in range(len(t)-1):
        # calculate y true
        y_t[j+1]=f_t(t[j+1])
        #calculate y_pred
        ## j+1=1
        if j+1==1:
          yp=y[j]+f(t[j], y[j])*dt #inisiasi yp
          yk = [y[j]+f(t[j], yp)*dt] # inisiasi y korektor
          e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100 # inisiasi error untuk y kor pertama
          while e[j+1]>er:
            yk.append(y[j]+f(t[j], yk[-1])*dt)
            e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100
          k[j+1]=len(yk)
          y[j+1]=yk[-1]

        ## j+1=2
        elif j+1==2:
          yp=y[j]+(3/2*f(t[j], y[j])-1/2*f(t[j-1], y[j-1]))*dt #inisiasi yp
          yk = [y[j]+(1/2*f(t[j], yp)+1/2*f(t[j], y[j]))*dt] # inisiasi y korektor
          e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100 # inisiasi error untuk y kor pertama
          while e[j+1]>er:
            yk.append(y[j]+(1/2*f(t[j], yk[-1])+1/2*f(t[j], y[j]))*dt)
            e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100
          k[j+1]=len(yk)
          y[j+1]=yk[-1]

        ## j+1=3
        elif j+1==3:
          yp=y[j]+(23/12*f(t[j], y[j])-16/12*f(t[j-1], y[j-1])\
                       +5/12*f(t[j-2], y[j-2]))*dt #inisiasi yp
          yk = [y[j]+(5/12*f(t[j], yp)+8/12*f(t[j], y[j])\
                       -1/12*f(t[j-1], y[j-1]))*dt] # inisiasi y korektor
          e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100 # inisiasi error untuk y kor pertama
          while e[j+1]>er:
            yk.append(y[j]+(5/12*f(t[j], yk[-1])+8/12*f(t[j], y[j])\
                       -1/12*f(t[j-1], y[j-1]))*dt)
            e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100
          k[j+1]=len(yk)
          y[j+1]=yk[-1]

        ## j+1=4
        elif j+1==4:
          #inisiasi yp
          yp=y[j]+(55/24*f(t[j], y[j])-59/24*f(t[j-1], y[j-1])\
                       +37/24*f(t[j-2], y[j-2])-9/24*f(t[j-3], y[j-3]))*dt
          #inisasi yk
          yk=[y[j]+(9/24*f(t[j], yp)+19/24*f(t[j], y[j])\
                       -5/24*f(t[j-1], y[j-1])+1/24*f(t[j-2], y[j-2]))*dt]
          #inisiasi e
          e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100
          #lakukan koreksi
          while e[j+1]>er:
            yk.append(y[j]+(9/24*f(t[j], yk[-1])+19/24*f(t[j], y[j])\
                       -5/24*f(t[j-1], y[j-1])+1/24*f(t[j-2], y[j-2]))*dt)
            e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100
          #simpan nilai
          k[j+1]=len(yk)
          y[j+1]=yk[-1]

        ## j+1>= 5
        else:
          #inisiasi yp
          yp=y[j]+(191/72*f(t[j], y[j])-281/72*f(t[j-1], y[j-1])\
                       +267/72*f(t[j-2], y[j-2])-131/72*f(t[j-3], y[j-3])+\
                       26/72*f(t[j-4], y[j-4]))*dt
          #inisiasi yk
          yk = [y[j]+(26/72*f(t[j], yp)+61/72*f(t[j], y[j])\
                       -21/72*f(t[j-1], y[j-1])+7/72*f(t[j-2], y[j-2])-\
                       1/72*f(t[j-3], y[j-3]))*dt]
          #inisiasi e
          e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100
          #lakukan koreksi
          while e[j+1]>er:
            yk.append(y[j]+(26/72*f(t[j], yk[-1])+61/72*f(t[j], y[j])\
                       -21/72*f(t[j-1], y[j-1])+7/72*f(t[j-2], y[j-2])-\
                       1/72*f(t[j-3], y[j-3]))*dt)
            e[j+1]=(np.abs(yk[-1]-y_t[j+1])/y_t[j+1])*100
          #simpan nilai
          k[j+1]=len(yk)
          y[j+1]=yk[-1]

    df = pd.DataFrame(list(zip(t, y, y_t, e, k)),
                  columns =['t','y pred', 'y true', 'error(%)', 'k'])
    if pr == 1:
      return y_t, y, t
    else:
      display(df)
      plt.plot(t, y, 'k', label='Euler Method')
      plt.plot(t, y_t, '0.7', label='Analytical')
      plt.title('Comparison For Euler Method and Analytical')
      plt.xlabel('t')
      plt.ylabel('f(t)')
      plt.grid()
      plt.legend(loc='lower right')
      plt.show()
```
